***ALGORITMA DECISION TREE dipakai untuk masalah klasifikasi, Kelebihan metode ini adalah kemampuannya untuk memilah aturan keputusan menuju ke keputusan yang diinginkan dalam bentuk pohon keputusan***

Import Library

membagi Menjadi data training dan data test

```
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import train_test_split
```

Data_Column = mendeklarasikan data_column dimana berisi kolom dari dataset haber yang nantinya dengan perintah .values diubah kebentuk array.

Data Class = mendeklarasikan algoritma data_class dimana berisi label/kelas dari dataset haber yang nantinya dengan perintah .values diubah kebentuk array.

```
data_column = dataset[['age','deg-malig']].values

data_class = dataset[['Class']].values
```

Mulai membagi data

```
(train_inputs, test_inputs, train_classes, test_classes) = train_test_split(data_column, data_class, train_size=0.8, random_state=1)
```

memanggil algoritma yang disimpan dalam variabel dtc.

```
dtc = DecisionTreeClassifier()
```

melakukan proses training / pelatihan

```
dtc.fit(train_inputs, train_classes)
```

menampilkan hasil akurasi

```
dtc.score(test_inputs, test_classes)
```

Predict Test Set Labels

```
y_pred = dtc.predict(test_inputs)
```

Menunjukkan presisi, perolehan, dan skor f1 metrik klasifikasi utama pada basis per kelas.

```
from sklearn.metrics import classification_report
print(classification_report(test_classes, y_pred))
```
